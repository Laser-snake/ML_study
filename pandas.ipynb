```python
def getPage(page = 0,text='Разработчик'):
    """
    Создаем метод для получения страницы со списком вакансий.
    Аргументы:
        page - Индекс страницы, начинается с 0. Значение по умолчанию 0, т.е. первая страница
    """
    
    # Справочник для параметров GET-запроса
    params = {
        'text': text, # Текст фильтра. В имени должно быть слово "Аналитик"
        'area': 1, # Поиск ощуществляется по вакансиям города Москва
        'page': page, # Индекс страницы поиска на HH
        'per_page': 100, # Кол-во вакансий на 1 странице
        #'only_with_salary':'True'
        'search_field':'name'
    }
    
    
    req = requests.get('https://api.hh.ru/vacancies', params) # Посылаем запрос к API
    items = req.json()['items'] # Декодируем его ответ, чтобы Кириллица отображалась корректно
    req.close()
    return items
```

```python
def explode(x):
    values = (sum([list(y.values()) if type(y) == dict else [y] for y in x ],[]))
    index = sum([list(map(lambda z: y+'.'+z,list(x[y].keys())))if type(x[y])==dict else [y] for y in x.index],[])
    return pd.Series(values,index=index)
```

```python
def get_language(x):
    import re
    return re.findall('go|python|c[#+]{,2}',x.lower())
```

```python
df1.assign(lang=df1.name.apply(lambda x: get_language(x))).explode('lang').\
groupby(['lang']).agg({'salary.from':'mean','salary.to':'mean','id':'nunique'})
```

```python
def get_qual(x):
    import re
    lst = re.findall('junior|middle|senior',x.lower())
    if len(lst) ==0:
        return['middle']
    else:
        return lst
```

```python
from sklearn.preprocessing import LabelEncoder

df2 = df1[['name','salary.from','snippet.requirement']].\
assign(lang=df1.name.apply(lambda x: get_language(x))).\
assign(qual=df1.name.apply(lambda x: get_qual(x)))[['salary.from','qual','lang']].fillna(170000)
```
