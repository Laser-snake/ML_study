```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = load_wine()['data']
y = load_wine()['target']
labels = load_wine()['feature_names']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from sklearn.ensemble import GradientBoostingClassifier


gbc = GradientBoostingClassifier(n_estimators=50,learning_rate=0.5,max_depth=1)
gbc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

gbc_tr = accuracy_score(y_train,gbc.predict(X_train))
gbc_test = accuracy_score(y_test,gbc.predict(X_test))
```

```python
print(gbc_tr,'   ',gbc_test)
```

```python
datAcc = {'name':[], 'train_accuracies':[], 'test_accuracies':[]}
dfAcc = pd.DataFrame(datAcc)
```

```python
dfAcc
```

```python
dfAcc.loc[len(dfAcc.index)] = ['GradientBoostingClassifier', gbc_tr, gbc_test]
```

```python
dfAcc
```

```python
from xgboost import XGBClassifier
```

```python
xgbc = XGBClassifier(n_estimators=50,learning_rate=0.5,max_depth=1)
xgbc.fit(X_train, y_train)
xgbc_tr = accuracy_score(y_train,xgbc.predict(X_train))
xgbc_test = accuracy_score(y_test,xgbc.predict(X_test))
print(xgbc_tr,'   ',xgbc_test)
```

```python
dfAcc.loc[len(dfAcc.index)] = ['XGBClassifier', xgbc_tr, xgbc_test]
dfAcc
```

```python
from lightgbm import LGBMClassifier

lgbmc = XGBClassifier(n_estimators=50,learning_rate=0.5,max_depth=1)
lgbmc.fit(X_train, y_train)
lgbmc_tr = accuracy_score(y_train,lgbmc.predict(X_train))
lgbmc_test = accuracy_score(y_test,lgbmc.predict(X_test))
print(lgbmc_tr,'   ',lgbmc_test)
dfAcc.loc[len(dfAcc.index)] = ['LGBMClassifier', lgbmc_tr, lgbmc_test]
```

```python
df = pd.read_excel('/content/default of credit card clients.xls',header=1)
X=df.iloc[:,1:-1]
y = df.iloc[:,-1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
methods = ['GradientBoostingClassifier','XGBClassifier','LGBMClassifier','CatBoostClassifier']
datAcc = {'name':[], 'train_accuracies':[], 'test_accuracies':[]}
dfAcc = pd.DataFrame(datAcc)

for method in methods:
  if method == 'GradientBoostingClassifier':
    boost = GradientBoostingClassifier(n_estimators=50,learning_rate=0.5)
    boost.fit(X_train, y_train)
    boost_tr = accuracy_score(y_train,boost.predict(X_train))
    boost_test = accuracy_score(y_test,boost.predict(X_test))
    dfAcc.loc[len(dfAcc.index)] = [method,boost_tr, boost_test]
  elif method == 'XGBClassifier':
    boost = XGBClassifier(n_estimators=100)
    boost.fit(X_train, y_train)
    boost_tr = accuracy_score(y_train,boost.predict(X_train))
    boost_test = accuracy_score(y_test,boost.predict(X_test))
    dfAcc.loc[len(dfAcc.index)] = [method,boost_tr, boost_test]
  elif method == 'LGBMClassifier':
    boost = LGBMClassifier(n_estimators=100)
    boost.fit(X_train, y_train)
    boost_tr = accuracy_score(y_train,boost.predict(X_train))
    boost_test = accuracy_score(y_test,boost.predict(X_test))
    dfAcc.loc[len(dfAcc.index)] = [method,boost_tr, boost_test]
  else:
    boost = CatBoostClassifier(n_estimators=100)
    boost.fit(X_train, y_train)
    boost_tr = accuracy_score(y_train,boost.predict(X_train))
    boost_test = accuracy_score(y_test,boost.predict(X_test))
    dfAcc.loc[len(dfAcc.index)] = [method,boost_tr, boost_test]


dfAcc
```
